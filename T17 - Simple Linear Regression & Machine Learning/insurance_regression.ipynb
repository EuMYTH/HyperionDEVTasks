## Practical Task 1
The purpose of this notebook is to analyze the impact of age on insurance costs and to generate predictive models based on this analysis.

```python
# Importing Libraris
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Reading the insurance data from a CSV file into a DataFrame
df = pd.read_csv("insurance.csv")

# Displaying the first few rows of de DataFrame for inspections
df.head()
```

```python
# Displaying information about the DataFrame, including data types and missing values
df.info()
```

```python
# Assigning column "age" to x variable and "charges" to y variable for future use
x = df.loc[:, "age"].values
y = df.loc[:, "charges"].values

# Reshaping the arrays to be 2-dimensional for compatibility with scikit-learn
x = x.reshape(-1, 1)
y = y.reshape(-1, 1)

# Creating a scatter plot of 'age' and 'charges' to visualise  the relationship between age and insurance charges
plt.scatter(x, y, color = "red")
plt.title("Age vs Charges")
plt.xlabel("Age")
plt.ylabel("Charges")
plt.show()
```

```python
# Creating an instance of the LinearRegression model
agecharges_model = LinearRegression()

# Fitting the model using "age" as input(x) and "charges" as target(y)
agecharges_model.fit(x, y)

# Generating predicted values for "charges" based on the input "age"
y_predict = agecharges_model.predict(x)

# Creating a scatter plot of the actual data points and a line plot of the predict values
plt.scatter(x, y, color = "red")
plt.plot(x, y_predict, color = "blue", label='Predicted Data')
plt.title("Predictions on Age vs Charges")
plt.xlabel("Age")
plt.ylabel("Charges")
plt.legend()
plt.show()
```
